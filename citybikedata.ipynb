```python
import pandas as pd
```

```python
# Create a new Gender Only dataframe for each month

# Change values in columns to identify Male and Female users. (January.csv)

jan = pd.read_csv('csv/january.csv')
genderjan = jan[["gender"]].copy()
cleanjan = genderjan.rename(columns={"gender": "gender_january"})
cleanjan['gender_january'] = cleanjan['gender_january'].replace({1: "Male", 2: "Female", 0: "Unknown"})
cleanjan
```

```python
# Change values in columns to identify Male and Female users. (february.csv)

feb = pd.read_csv('csv/february.csv')
genderfeb = feb[["gender"]].copy()
cleanfeb = genderfeb.rename(columns={"gender": "gender_february"})
cleanfeb['gender_february'] = cleanfeb['gender_february'].replace({1: "Male", 2: "Female", 0: "Unknown"})
cleanfeb
```
